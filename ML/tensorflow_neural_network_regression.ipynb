```python
tf.random.set_seed(42)
model=tf.keras.Sequential(
    [tf.keras.layers.Dense(1)
])
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(),loss=tf.keras.losses.MAE,metrics=["mae"])
```

```python
model.fit(tf.expand_dims(X,axis=-1),y,epochs=5)
```

```python
model.predict(np.array([[17.0]]))
```

```python
X=tf.range(-100,100,4)
X
```

```python
y=X+10
y
```

```python
plt.scatter(X,y)
```

```python
X_train=X[:40]
y_train=y[:40]
X_test=X[40:]
y_test=y[40:]
X_test
```

```python
plt.figure(figsize=(10,7))
plt.scatter(X_train,y_train,c="b",label="Training Data")
plt.scatter(X_test,y_test,c="g",label="Testing Data")
plt.legend()
```

```python

model3=tf.keras.Sequential(name="ModelTest")
model3.add(tf.keras.Input(shape=[1]))
model3.add(tf.keras.layers.Dense(10,name="Input_layer"))
model3.add(tf.keras.layers.Dense(100,name="Hidden_layer"))
model3.add(tf.keras.layers.Dense(1,name="Output_layer"))

```

```python
model3.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),loss=tf.keras.losses.MAE,metrics=["mae"])
```

```python
#X_train,X_test
```

```python
model3.summary()
```

```python
#from re import VERBOSE
#X_train = np.array(X_train)
#X_train = tf.expand_dims(X_train,-1)
#tf.expand_dims(X_train,axis=-1).shape,y_train.shape
model3.fit(X_train,y_train,epochs=100,verbose=1)
```

```python
from tensorflow.keras.utils import plot_model
plot_model(model3,show_shapes=True)
```

```python
y_predict=model3.predict(X_test)
y_predict
#X_test
```

```python
y_test
```

```python
def plot_predictions(train_data=X_train,
                     train_labels=y_train,
                     test_data=X_test,
                     test_labels=y_test,
                     predictions=y_predict):
  plt.figure(figsize=(10,7))
  plt.scatter(train_data,train_labels,c="b",label="Training Data")
  plt.scatter(test_data,test_labels,c="g",label="Testing Data")
  plt.scatter(test_data,predictions,c="r",label="Testing Data")
  plt.legend()
```

```python
plot_predictions()
```
